## KL Divergence

En Teoría de Probabilidades, y Teoría de la información, la **Divergencia de Kullback - Leibler** es una medida no simétrica de la similitud o diferencia entre dos funciones de distribución de probabilidad $P$ y $Q$. Generalmente, $P$ representa la verdadera distribución de los datos, y $Q$ representa un modelo o aproximación de $P$.

Para distribuciones de probabilidad de una variable continua, su divergencia KL se define como:

$$ D_{KL}(P||Q) = \int_{-\infty}^{\infty} p(x)\ln \dfrac{p(x)}{q(x)} dx $$

donde $p$ y $q$ representan las densidades de $P$ y $Q$, respectivamente. Es posible, alternativamente, escribir la expresión para una variable discreta:

$$ D_{KL}(P||Q) = \sum_{i} p(i)\ln \dfrac{p(i)}{q(i)} $$

#### Forward and Reverse

Típicamente, a la divergencia KL calculada como $D_{KL}(P||Q)$, siendo $P$ la distribución original y $Q$ el modelo que aproxima. Debido a que  $D_{KL}(P||Q) \neq  D_{KL}(Q||P)$ (no hay simetría), a la primera se le suele llamar **Divergencia KL Forward**, y a la segunda **Divergencia KL Reverse**. En lo que viene, estudiaremos sus diferentes interpretaciones para una distribución $P$ bimodal.


```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
from scipy.integrate import quad, dblquad
from scipy.optimize import minimize
from scipy.special import gamma
from scipy.stats import multivariate_normal
import matplotlib.mlab as mlab
%matplotlib inline
plt.rcParams["figure.figsize"]=(10,8)
```

```python
#Parametros relevantes

## Variables Globales

### Intervalos de Integración
a,c = 0,0
b,d = 9,9


### Parametros para distribución P bimodal. Son globales!
mu1, mu2 =  [3,3],[8,8]
cov1 , cov2 =[[3,0],[0,3]] , [[2,0],[0,2]]
w=0.5

### Parametros distribucion Q
qcov = [[2,0],[0,2]]
```

```python
##### Funciones de distribucion para calculos:


## Para evitar problemas, a cada distribución le damos los argumentos tales como media, covarianza, etc
def P(pos,args):
    ## pos: Par (x,y)
    ##args : mu1,mu2,cov1,cov2 para la distribución bimodal
    mu1=args[0]
    mu2=args[1]
    cov1=args[2]
    cov2 =args[3]
    w = args[4]
    return w*multivariate_normal.pdf(pos,mean=mu1,cov=cov1)+(1-w)*multivariate_normal.pdf(pos,mean=mu2,cov=cov2)

def Q(pos,args):
    ### Q es una normal, por lo tanto le damos argumentos
    #### pos : par (x,y)
    #### args: mu y cov
    mu = args[0]
    cov = args[1]
    return multivariate_normal.pdf(pos,mean=mu,cov=cov)

#### Función para Graficar


def GraficoCurvas(Ps,delta,args,col):
    
    # Ps : Distribucion a graficar
    # delta: salto en el arreglo
    # args: argumentos a entregar a la distribucion
    # col : color del grafico
    
    x = np.arange(-4, 18.0, delta)
    y = np.arange(-4, 18.0, delta)
    
    X, Y = np.meshgrid(x, y)
    pos = np.empty(X.shape + (2,))
    pos[:, :, 0] = X; pos[:, :, 1] = Y
    z = Ps(pos,args)
    plt.contour(X, Y,z,10,colors=col)
    plt.xlim(-4,18)
    plt.ylim(-4,18)
    plt.title('Gráfico Curvas de Nivel')

```

```python
### KL Divergence : Aquí utilizamos las pdf de P y Q

def KLDivergenciaForward(P,Q,argsP,argsQ,delta,a=a,b=b,c=c,d=d):
    
    # P: distribucion a ser aproximada
    # Q: distribucion Test que aproxima a P
    # argsP: argumentos de P
    # argsQ: argumentos de Q
    # a,b : intervalos de integración para x
    # c,d : intervalos de integración para y
    
    ## Integrando: Separamos logaritmo simplemente para evitar problemas
    
    Integrando = lambda point : P(point,argsP)*np.log(P(point,argsP))-P(point,argsP)*np.log(Q(point,argsQ))
    
    ### Array Creation
    
    x = np.arange(a, b, delta)
    y = np.arange(c, d, delta)
    X, Y = np.meshgrid(x, y)
    pos = np.empty(X.shape + (2,))
    pos[:, :, 0] = X; pos[:, :, 1] = Y
     
    ### Iteracion KL - Divergence
    KL = 0
    for segmento in pos:
        for point in segmento:
            KL += Integrando(point)
    return KL

def KLDivergenciaReverse(P,Q,argsP,argsQ,delta,a=a,b=b,c=c,d=d):
    
    # P: distribucion a ser aproximada
    # Q: distribucion Test que aproxima a P
    # argsP: argumentos de P
    # argsQ: argumentos de Q
    # a,b : intervalos de integración para x
    # c,d : intervalos de integración para y
    
    ## Integrando: Separamos logaritmo simplemente para evitar problemas
    
    Integrando = lambda point : Q(point,argsQ)*np.log(Q(point,argsQ))-Q(point,argsQ)*np.log(P(point,argsP))
    
    ### Array Creation para calulcar KL divergencia
    
    x = np.arange(a, b, delta)
    y = np.arange(c, d, delta)
    X, Y = np.meshgrid(x, y)
    pos = np.empty(X.shape + (2,))
    pos[:, :, 0] = X; pos[:, :, 1] = Y  ### Con esto obtenemos un array de puntos (x,y) que definen la grilla
     
    ### Iteracion KL - Divergence
    KL = 0
    for segmento in pos:
        for point in segmento:
            KL += Integrando(point)
    return KL

```

Como el objetivo es encontrar una distribución $Q$ que minimize la ``distancia``a $P$, implementamos métodos de optimización que minimicen con respecto a la media de la distribución Q.

```python
def OptimizarForward(P,Q,argsP,covq,delta,x0):  
    
    # P: distribucion a ser aproximada
    # Q: distribucion Test que aproxima a P
    # argsP: argumentos de P
    # covq : Covarianza para Q. Recordar que estamos buscando la media de Q que minimiza la divergencia
    # delta: tamaño del paso
    # x0: Para los métodos de optimizazión, punto inicial.
    
    funcObj = lambda qmu : KLDivergenciaForward(P,Q,argsP,(qmu,covq),delta,a,b,c,d)
    optimo = minimize(funcObj,x0,options={"maxiter":40}).x
    return optimo

def OptimizarReverse(P,Q,argsP,covq,delta,x0):
    
    # P: distribucion a ser aproximada
    # Q: distribucion Test que aproxima a P
    # argsP: argumentos de P
    # covq : Covarianza para Q. Recordar que estamos buscando la media de Q que minimiza la divergencia
    # delta: tamaño del paso
    # x0: Para los métodos de optimizazión, punto inicial.
    
    funcObj = lambda qmu : KLDivergenciaReverse(P,Q,argsP,(qmu,covq),delta,a,b,c,d)
    optimo = minimize(funcObj,x0,options={"maxiter":40}).x
    return optimo  
```

#### Ejecución

Comenzamos con los parámetros necesarios para encontrar la distribución $Q$.

```python
## Parametros para la ejecucion
x0 = [5,5]
argsP = (mu1,mu2,cov1,cov2,w) ## Argumentos a entregar a P
delta1 = 0.025 # StepSize para ploteo
delta2 = 0.65 # Stepsize para calculo minimo
```

A continuación, encontramos la media para $Q$ que minimiza **KLD Forward** y **KLD Reverse**, respectivamente.

```python
### Minimizacion KLDivergence Forward
optimoForward = OptimizarForward(P,Q,argsP,qcov,delta2,x0)
### Minimizacion KLDivergence Reverse
optimoReverse = OptimizarReverse(P,Q,argsP,qcov,delta2,x0)
```

Luego, "empaquetamos" los argumentos para ser entregados a la distribución $Q$ y finalmente ser graficados

```python
### Argumentos Distribiciones
argsQForward=(optimoForward,qcov)
argsQReverse=(optimoReverse,qcov)
```

```python
### Grafico de Distribuciones

GraficoCurvas(P,delta1,argsP,"b")       # Grafico P 
GraficoCurvas(Q,delta1,argsQReverse,"r")# Q : KLD Reverse
GraficoCurvas(Q,delta1,argsQForward,"g")# Q : KLD Forward
```

Como es posible ver en el gráfico de curvas de nivel, **KLD Reverse** tiene a acercarse a la zona de mayor masa de la distribución, es decir, el foco con mayor varianza. Por el contrario, **KLD Forward** intenta tener mayor probabilidad de ocurrencia, y por lo tanto, intenta "agarrar" lo más posible la distribución.

## Mean Field Approximation

Si se quieren estimar los parametros $\mu$, $\theta$ de una distribución Normal desconocida, a partir de datos conocidos. A partir del enfoque **Mean Field**, aproximamos con $q(\mu,\sigma)$ como el producto de distribuciones indepentiendes:

$$ q(\mu,\sigma) = q_{\mu}(\mu) q_{\sigma}(\sigma)$$

Donde, asumiendo $\lambda_0=\mu_0=a_0=b_0=0$:
   1. La distribución de $q_{\mu} \sim N(\mu_N,\lambda_N^{-1})$, donde:
   
       a. $\mu_N = \frac{\lambda_0 \mu_0 + N \bar{x}}{\lambda _0 + N}=\bar{x}$
       
       b. $\lambda_N= \mathbb{E}_{q_{\sigma}}(\sigma) (\lambda_0 + N) = N \mathbb{E}_{q_{\sigma}}(\sigma)$
       
   2. La distribución de $q_{\sigma} \sim \text{Gamma} (a_N, b_N)$, donde:

    a) $a_N = a_0 + \frac{N}{2} = \frac{N}{2}$
    
    b) $b_N = b_0 + \frac{1}{2} \mathbb{E}_{q_{\mu}} \left[ \sum_{n=1}^N (x_n - \mu)^2 + \lambda_0 (\mu - \mu_0) \right] =\frac{a_N}{\mathbb{E}_{q_{\sigma}}(\sigma)}$
    
Las esperanzas mencionadas para $\mu$ son :
$$\mathbb{E}_{q_{\mu}}(\mu) = \bar{x}\\  
\mathbb{E}_{q_{\mu}}(\mu ^2) = (\bar{x})^2 + \frac{1}{N\mathbb{E}_{q_{\sigma}}(\sigma)}$$

Para ver la iteración, utilizando lo anterior notamos que :
$$ \dfrac{1}{\mathbb{E}_{q_{\sigma}}(\sigma)}= \bar{x^2} - 2\bar{x} \mathbb{E}_{q_{\mu}}(\mu) + \mathbb{E}_{q_{\mu}} (\mu^2)\\
=\bar{x^2} - 2(\bar{x})^2+(\bar{x})^2 + \frac{1}{N\mathbb{E}_{q_{\sigma}}(\sigma)} \\
= \bar{x^2} - (\bar{x})^2 + \frac{1}{N\mathbb{E}_{q_{\sigma}}(\sigma)} $$

Y por lo tanto:
$$(\mathbb{E}_{q_{\sigma}}(\sigma))_{i+1} =  \left( \bar{x^2} - (\bar{x})^2 + \frac{1}{N(\mathbb{E}_{q_{\sigma}}(\sigma))_i} \right)^{-1}$$
Así, podemos calcular iterativamente $(\mathbb{E}_{q_{\sigma}}(\sigma))_{i\in\mathbb{N}}$, partiendo inicialmente con $(\mathbb{E}_{q_{\sigma}}(\sigma))_0 =1$. En cada iteración generamos una tupla $(a_N,b_N,\mu_N,\lambda_N)_i$ y por lo tanto, una distribución $Q_i$

Es posible establecer criterios de parada. Debido a que la formulación anterior tiene forma de **punto fijo**, fijamos un número de iteraciones máxima, pues no tenemos seguridad en la convergencia de $Q_i$ a $P$ (donde $Q_i$ es la distribución generada en la iteración $i$).

```python
#Parámetros

mu,sigma = 0,1
N= 100
D = np.random.normal(mu,sigma,N)   
maxiter = 50
mean = np.mean(D)
vol = 0.1

a ,b = mean-0.05,mean+0.05
c,d = 0.6,1.3

## Parametros para P

argsP=(N,D)
```

```python
## Calculos Varios
def E_Q_mu(D=D):
    return np.mean(D)
def E_Q_mu2(EspSigma,D=D,N=N):
    return E_Q_mu(D)+EspSigma/N
def E_Q_sigma(EspSigmaAnterior,D=D,N=N):
    return (np.mean(D**2)-np.mean(D)*E_Q_mu(D)+(1/(N*EspSigmaAnterior)))**(-1)
    
## Densidades

def P(pos,args):
    ## Entradas
    N=args[0]
    D=args[1]
    mu = pos[0]
    sigma = pos[1]
    
    ## Parametros
    lamb = N**2
    alpha = N/2
    m = np.mean(D)
    beta = np.sum(D**2)/2
    
    #Resultado 
    return ((beta**alpha)*np.sqrt(lamb)/gamma(alpha)*(np.sqrt(2*np.pi)))*(sigma**(alpha-0.5))*np.exp(-1*beta*sigma)*np.exp(-0.5*(lamb*sigma*((mu-m)**2)))

def Gamma(theta,a,b):
    return (b**(a))*(theta**(a-0.5))*(np.exp(-b*theta))/(gamma(a)) 
```

```python
### Iteración para encontrar la distrubución Q

def MeanField(P,Q,delta,maxiter,KLs):
    EsperanzaMu = E_Q_mu(D)
    EsperanzaSigma = 1.2
    
    counter = 0
    while counter < maxiter:
        ## Calculo Qmu y Qsigma
        an = N/2
        bn = an/EsperanzaSigma
        
        Q_sigma = lambda sigma,m,n: Gamma(sigma,m,n)
        
        ### Parametros Gaussiana
        
        muN = EsperanzaMu
        lambdaNInv = (N*EsperanzaSigma)**(-1)
                
        Q_mu = lambda mu, loc,scale: stats.norm.pdf(mu,loc=loc,scale=scale)
        
        ### Definicion Q
        
        Q = lambda pos,args : Q_mu(pos[0],args[0],args[1])*Q_sigma(pos[1],args[2],args[3])
        
        ### Empaquetamiento argumentos para Q
        
        argsQ = (muN,lambdaNInv,an,bn)
               
        ## Calculo divergencia
        
        dist = KLs(P,Q,argsP,argsQ,delta,a,b,c,d)
        EsperanzaSigma = E_Q_sigma(EsperanzaSigma)
        
        counter +=1
        
    return Q,argsQ,dist,counter
        
    
```

Tras cierto número de iteraciones, recuperamos las distribuciones.

```python
QForward,argsQForward,distF,counterF=MeanField(P,Q,delta2,maxiter,KLDivergenciaForward)
QReverse,argsQReverse,distR,counterR=MeanField(P,Q,delta2,maxiter,KLDivergenciaReverse)
```
