# Tarea 1: Diagnóstico de Convergencia en Cadenas de Markov

Para poder realizar inferencias validas, generando *samples*  o cadenas a través de algoritmos *Markov Chain Monte Carlo* (MCMC), se basa en el supuesto de que dichas cadenas provienen de una distribución estacionaria, o distribución posterior verdadera de interés. La teoría, en concreto las *Leyes de los grande números*, aseguran que en esperanza la condición de convergencia a la distribución buscada se obtiene cuando el número de iteraciones tiende a infinito. Puesto que no es factible calcular infinitas muestras, es necesario fijar un número minimo de iteraciones bajo el cual se obtenga una aproximación razonable a la distribución objetivo.

Sin embargo, no existe un umbral universal bajo el cual asegurar convergencia. Por esto, se debe hacer un diagnóstico individual cada vez que una cadena de Markov es generada. Dichos diagnósticos son conocidos como **diagnósticos de convergencia**.

De todas formas, nunca ppodemos garantizar convergencia. Por lo tanto, los diagnósticos presentados son test estadísticos que permiten entregar información de convergencia, desde un punto de vista estadístico y no absoluto. 

En la siguiente tarea se presentan definiciones básicas, código e implementación para métodos vistos en clases, de tres técnicas para evaluar la convergencia de Cadenas de Markov.

Las técnicas de diagnóstico a estudiar son:
   1. Gelman - Rubin diagnostic
   2. Visualización de valor esperado
   3. Geweke diagnostic


# Algoritmo Programado en Clases : Metropolis-Hastings

A continuación, mostramos una implementación del algoritmo de Metropolis-Hastings (MH).


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from scipy.integrate import quad
from scipy.misc import factorial
%matplotlib inline
```

A continuación, la funciones de distribución $Q$ de kernel, y la distribución de la cuál se busca generar una muestra $P^*$.

```python
def Ps1(x):
    p = x*(1-x)
    return p
def Ps2(x):
    p = x**3*(1-x)
    return p
def Qgen(xprev,xmin,xmax,my_std,size=1): #Para generar variables aleatorias
    myclip_a = xmin
    myclip_b = xmax
    my_mean = xprev
    a, b = (myclip_a - my_mean) / my_std, (myclip_b - my_mean) / my_std
    return stats.truncnorm.rvs(a,b,loc=my_mean,scale=my_std,size=size)
def Qdist(x,xprev,xmin,xmax,my_std): # calcular pdf
    myclip_a = xmin
    myclip_b = xmax
    my_mean = xprev
    a, b = (myclip_a - my_mean) / my_std, (myclip_b - my_mean) / my_std
    return stats.truncnorm.pdf(x,a,b,loc=my_mean,scale=my_std)
```

Gráfico de todas las funciones

```python
plt.figure(figsize=(3,2))
x = np.linspace(0,1)
plt.plot(x,Ps1(x))
plt.plot(x,Ps2(x))
```

Además, el algoritmo de Metropolis-Hastings.

```python
def MetropolisHastings(x0,k,Qdist,Qgen,Ps,sigma):
    samples = []
    xs = x0
    for i in range(k):
        x = xs
        xp = Qgen(x,0,1,sigma,1)[0]
        alpha = (Ps(xp)*Qdist(x,xp,0,1,sigma))/(Ps(x)*Qdist(xp,x,0,1,sigma))
        #alpha =Ps(xp)/Ps(x)
        r = min(1,alpha)
        u = np.random.uniform(0,1)
        if u<r:
            xs = xp
        else:
            xs = xs
        samples.append(xs)
    return (samples)
def CompararGraficamente(x,Ps):
    plt.hist(x,normed=True,bins=50)
    plt.plot(np.linspace(0,1),Ps(np.linspace(0,1))/quad(Ps,0,1)[0])
```

Parametros de control

```python
k = 1000
x0 = 0.1
sigma = 0.1
```

```python
x1= MetropolisHastings(x0,k,Qdist,Qgen,Ps1,sigma)
CompararGraficamente(x1,Ps1) # Histograma vs Funcion de distribución
```

```python
x2= MetropolisHastings(x0,k,Qdist,Qgen,Ps2,sigma)
CompararGraficamente(x2,Ps2) # Histograma vs Funcion de distribución
```

```python
k = 100000
x1= MetropolisHastings(x0,k,Qdist,Qgen,Ps1,sigma)
CompararGraficamente(x1,Ps1) # Histograma vs Funcion de distribución
```

```python
x2= MetropolisHastings(x0,k,Qdist,Qgen,Ps2,sigma)
CompararGraficamente(x2,Ps2) # Histograma vs Funcion de distribución
```

Vemos que la aproximación, en este caso, es mucho mejor que antes, para ambas distribuciones. Sin embargo, aún así hay errores. Es necesario un número **mucho mayor** de iteraciones para alcanzar un error despreciable.

## Gelman-Rubin Convergence ratio

Este test evalua la convergengia de una MCMC a través del análisis de varias cadenas de Markov. Para ello, se basa en un análisis de varianza para hablar sobre la convergencia. Esto es, calcular varianzas **Between - Chain** (varianza entre las cadenas) y **Within-Chain** (varianza con respecto a si misma). Grandes diferencias entre estas cantidades indican ausencia de convergencia.

Un algoritmo estándar es (ver cita):

   1. Generar $m\geq2$ cadenas de largo $2n$, para puntos iniciales lejanos. En nuestro caso, las generamos con el algoritmo de Metropolis-Hastings presentado al inicio.
   2. Descartar los primeros $n$ generados.
   3. Calcular las varianzas **Within-Chain** y **Between - Chain**. Si se tienen $m$ cadenas, cada una de $n$ elementos, calculamos:
       $$W = \dfrac{1}{m}\sum_{j=1}^m s_j^2 $$
       con $s_j^2 = \dfrac{1}{n-1}\sum_{i=1}^n(\theta_{ij}-\bar \theta_j)^2$ es la varianza estimada de los datos de la $j$-ésima cadena. Así, $W$, también llamada es el promedio de las varianzas de cada cadena. Por otro lado, calculamos:
       $$ B = \dfrac{n}{m-1} \sum_{j=1}^m(\bar \theta_j-\bar{\bar{ \theta}})^2$$
      donde :
      $$ \bar{\bar{ \theta}} = \dfrac{1}{m}\sum_{j=1}^m\bar\theta_j$$
     es simplemente un "promedio de promedios". $B$ representa la varianza de las medias de las cadenas, multiplicada por $n$ puesto que cada cadena está basada en $n$ realizaciones del experimento.
   4. Calcular la varianza total del parametro como una suma ponderada de las dos varianzas mencionadas. Con las definiciones anteriores, podemos estimar (insesgadamente) la varianza de la distribución estacionaria de la cadena:
   $$\hat{Var}(\theta) = (1-\dfrac{1}{n})W+\dfrac{1}{n}B$$
   5. Calcular el **Potential Scale reduction Factor**, definido como:
   $$ \hat{R} = \sqrt{\dfrac{ \hat{Var}(\theta)  }{W}}$$
   Intuitivamente, un valor alto de $\hat{R}$ (por ejemplo, mayor que 1.1 o 1.2), indica que la cadena se debe correr un mayor número de iteraciones para obtener una mejor convergencia a la distribución estacionaria.

A continuación presentamos una función que recibe una lista de cadenas, y entrega el **Potential Scale Reduction Factor**.

```python
def GelmanRubin(mcmc):
    m = len(mcmc)
    n = len(mcmc[0])
    mean = [np.mean(np.array(i)) for i in mcmc]
    Mean = np.mean(mean)
    s = [ (np.sum((i - np.mean(np.array(i)))**2)/(len(i)-1))**2 for  i in mcmc]
    W = np.mean(s)
    B = (n/(m-1))*np.sum((mean-Mean)**2)
    var = W - W/n +B/n
    R = np.sqrt(var / W)
    return R
        
```

Luego, generamos muestras repitiendo el algoritmo. Recordamos que es necesario eliminar la primera mitad de las muestras. Además, es necesario partir de puntos distintos. Para todos usamos la misma distribución $Q$ del algoritmo MH.

## Visualización de Media

Una técnica de diagnóstico util, en primera instancia, es analizar como la media de la cadena converge a la media de la distribución buscada.



```python
def medias(sample):
    medias = []
    iterates = []
    for i in range(len(sample)):
        test = sample[:i+1]
        medias.append(np.mean(test))
        iterates.append(i)
    return iterates,medias
def Visualizar(sample,Ps):
    esperanza = lambda x: x*Ps(x)
    mediaDist = quad(esperanza,0,1)[0]/quad(Ps,0,1)[0]
    iterates, means = medias(sample)
    plt.plot(iterates,means)
    plt.plot((np.min(iterates), np.max(iterates)), (mediaDist, mediaDist), '--')
    return 
```

Generamos muestras aleatorias de tamaño fijo, con distinto punto inicial:

```python
k=2000
y1= np.array(MetropolisHastings(0.1,2*k,Qdist,Qgen,Ps1,0.1))
y2= np.array(MetropolisHastings(0.8,2*k,Qdist,Qgen,Ps1,0.1))
y3= np.array(MetropolisHastings(0.5,2*k,Qdist,Qgen,Ps1,0.1))
```

```python
Visualizar(y1,Ps1)
```

```python
Visualizar(y2,Ps1)
```

```python
Visualizar(y3,Ps1)
```

Vemos que la primera y la segunda muestra se acercan mucho a la media de la distribución, pero la última muestra tan solo una tendencia a la media. Para esta última, son necesarias más iteraciones. Graficando todo junto:

## Geweke diagnostic

El test de **Geweke** compara la media y la varianza de dos segmentos o ventanas de una cadena de Markov individual(un porcentaje al inicio  y otro al fina de la cadenal). Para ello, se utiliza un test de diferencia de medias para evaluar la hipótesis nula de que ambas medias son iguales, y por lo tanto ambas partes de la cadena provienen de la misma distribución.

Se calcula un estadístico:

$$ z = \dfrac{\bar\theta_a-\bar\theta_b}{\sqrt{Var(\theta_a)+Var(\theta_b)}} $$

donde $a$ y $b$ representan las *ventanas* de la cadena. Se realiza un test de hipótesis, utilizando como hipótesis nula la mencionada anteriormente.


Para una mejor interpretación gráfica, se calculan varios estádisticos o **z-scores** tomando dos segmentos obtenidos al particionar la cadena en una cantidad finita de intervalos (extraídos de las distintas ventanas).  Si los **z-scores ** (que teóricamente están distribuidos normal estándar) de estos dos segmentos mencionados (el final y el inicial) son iguales, hay evidencia para decir que la cadena converge a una distribución estacionaria.

El código implementado calcula los **z-scores** de la diferencia entre varios segmentos iniciales a lo largo de la cadena, y el porcentaje restante de la cadena. Si la cadena ha convergido, la mayoría de los valores calculados se debe encontrar entre 2 desviaciones estándar del 0. Las funciones implementadas están **fuertemente** basadas en PyMC y su código fuente (ver referencias en Github).

```python
def geweke(x,first,last,intervals):
    x = np.array(x)
    zscores = []
    end = len(x) - 1

    sindices = np.arange(0, end/2, step = int((end / 2) / (intervals-1)))
    for start in sindices:
        # Calculate slices
        first_slice = x[int(start) : int(start) + int(first * (end - start))]
        last_slice = x[int(end - last * (end - start)):]
        
        z = (first_slice.mean() - last_slice.mean())
        z /= np.sqrt(first_slice.std()**2 + last_slice.std()**2)
        
        zscores.append([start, z])
    
    return zscores 
```

```python
def GraficarGeweke(zscores):
    plt.figure()
    x, y = np.transpose(zscores)
    plt.scatter(x.tolist(), y.tolist())

    plt.xlabel('First iteration', fontsize='x-small')
    plt.ylabel('Z-score ', fontsize='x-small')

    # Ejes máximos para las varianzas
    plt.plot((np.min(x), np.max(x)), (2, 2), '--')
    plt.plot((np.min(x), np.max(x)), (-2, -2), '--')

    plt.ylim(min(-5, np.min(y)), max(5, np.max(y)))
    plt.xlim(0, np.max(x))    
```

Si graficamos los **z-scores** para algunas cadenas generadas con el algoritmo de Metropolis-Hastings.

```python
zscores = geweke(x1,.1,.5,20)
GraficarGeweke(zscores)
zscores = geweke(x2,.1,.5,20)
GraficarGeweke(zscores)
zscores = geweke(x3,.1,.5,20)
GraficarGeweke(zscores)
```

La aparición de scores que estén dentro del umbral de 2 desviaciones estándar, hay evidencia estádistica de que las medias son iguales, y por tanto vienen de una misma distirbución estacionaria. Si los scores salen de dicho umbral, entonces se recomienda la realización de una mayor número de iteraciones. Vemos que los ejemplos 1 y 2 se salen del umbral, y por lo tanto es recomendable un mayor número de iteraciones, lo cual es consistente con lo observado anteriormente.
