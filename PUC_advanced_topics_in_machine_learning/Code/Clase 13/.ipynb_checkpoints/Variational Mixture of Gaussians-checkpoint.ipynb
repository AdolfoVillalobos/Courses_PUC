# Variational Mixture Of Gaussians

En esta tarea, se implementa la técnica "Variational Mixture Of Gaussians", desarrollada en el apunte del curso.

En este método, se calcula una distribución $q(Z,\pi,\mu,\Lambda) = q(Z)\cdot q(\pi) \cdot \Pi_{k=1}^Kq(\mu_k,\Lambda_k)$

donde :

1. $q(Z) = \Pi_{n=1}^n \Pi_{k=1}^K r_{nk}^{z_{nk}}$
2. $q(\pi) = Dirichlet(\pi | \alpha)$
3.$q(\mu_k,\Lambda_k) = N(\mu_k |m_k,(\beta_k\Lambda_k)^{-1})\cdot Wishart(\Lambda_k|W_k,v_k)$

De donde la derivación de las variables necesarias está hecha en detalle en el apunte,. y más adelante en las funciones. Debido a la interdependencia de las esperanzas, iniciamos con un valor inicial elegido arbitrariamente para la matriz $r = (r_{nk})_{n,k}$

Si se considera que $X = (x_n)$ son los datos, y las variables $\alpha_0, \beta_0, v_0$ y $m_0$ están dadas. Entonces:
1. $N_k = \sum_{n=1}^N r_{nk}$
2. $\bar{x}_k = \dfrac{1}{N_k} \sum_{n=1}N r_{nk}x_n$ 
3. $S_k = \dfrac{1}{N_k}\sum_{n=1}^N r_{nk}(x_n-\bar{x_k})(x_n-\bar{x_k})^T$
4. $\alpha_k = \alpha_0 + N_k$
5. $\beta_k = \beta_0 + N_k$
6. $v_k = v_0+N_k$
7. $W_k^{-1} = W_0^{-1}+N_kS_k+\dfrac{\beta_0N_k}{\beta_0+N_k}(\bar{x}_k-m_0)(\bar{x}_k-m_0)^T$
8. $r_{nk} = \dfrac{\rho_{nk}}{\sum_{j=1}^K \rho_{nj}}$
9. $\ln \rho_{nk} = \mathbb{E}(\ln \pi_k ) + 0.5\cdot \mathbb{E}(\ln | \Lambda_k|)- \dfrac{D}{2} \ln (2 \pi) - 0.5 \cdot (D \beta_k^{-1}+v_k(x_n-m_k)^TW_k(x_n-m_k))$

Adicionalmente, el cálculo de las esperanzas es:
1. $ \mathbb{E}(\ln \pi_k ) = \phi (\alpha_k) -\alpha(\hat{\alpha} ) $
2. $\mathbb{E}(\ln | \Lambda_k|) = \sum_{i=1}^D\phi(\dfrac{v_k+1-i}{2})+D \ln 2+\ln | W_k|$

donde la función $\phi$ es la función digamma.

En el código, se ha hecho un esfuerzo de vectorización, evitando iteraciones sobre la dimensión de los datos, o la cantidad de datos. Sin embargo, en algunos casos específicos se utilizó un for sobre la cantidad de Gaussianas. Entendemos que para un caso real de "very Big data", la cantidad de gaussianas puede ser muy alta. Sin embargo, para estos efectos, es poco infliuyente. Además, en aquellos casos en que se ha devido usar un for sobre la cantidad de clusters, los cálculos han sido hechos eficientemente.

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
from scipy.integrate import quad, dblquad
from scipy.optimize import minimize
from scipy.special import gamma,digamma,logsumexp
from scipy.stats import multivariate_normal,wishart,dirichlet
import matplotlib.mlab as mlab
import pandas as pd
%matplotlib inline
plt.rcParams["figure.figsize"]=(10,8)
```

```python
#### Declaración Base de datos

data = pd.read_csv('datos_clustering.txt', sep=",",header = None)
plt.plot(data[0],data[1],"o")
```

```python
matriz = data.values
D = data.shape[1]
N = data.shape[0]
```

```python
# Parametros 

K = 5 # Numero de Gaussianas
maxtol = 10
```

```python
# Hiperparametros

m0 = np.zeros((D,)).T
a0,b0,v0 = 0.001,1.0,1.0+D
W0 = K*np.eye(D)
```

```python
# Inicializaciones Vectoriales. Siempre pensando en logaritmos

rinit = (1.0/K)*np.random.random((N,K))
#rinit = (1/K)*np.random.rand(N,K)
#rinit = (1/K)*np.ones((N,K))
```

```python
#Funciones 

## Ecuaciones Escenciales

def phi(alpha):
    return digamma(alpha)

def E_lnPi(alpha):  ## 10.66
    return phi(alpha)-phi(np.sum(alpha))

def E_lnLambda(vt,Wt): ### Al lidiar con W_k, era mejor hacerlo con un for. De todas formas, se utilizan funciones bastante optimizadas como sum y listas comprimidas
    result = []
    for i in range(K):
        Wk = np.linalg.det(Wt[i])
        r = np.sum([phi(0.5*(vt[i]+1-(j+1))) for j in range(D)])+D*np.log(2.0)+np.log(Wk)
        result.append(r)
    return np.array(result)

def Esperanza_dificil(beta,vt,m,Wt): #10.64
    result = []
    for i in range(K):
        Xk = np.subtract(matriz,m[:,i]) # x_n-x_k
        result.append(vt[i]*((np.dot(Xk, Wt[i]) * Xk).sum(1))+D*beta[i]**(-1))
    return np.array(result)

def ln_rho(alpha,beta,vt,Wt,m): ## 10.46 
    rowvec = E_lnPi(alpha)+0.5*E_lnLambda(vt,Wt)-(D/2.0)*np.log(2*np.pi)
    lnrho = np.tile(rowvec,(N,1)).T  
    return (lnrho-0.5*Esperanza_dificil(beta,vt,m,Wt))
    
def R_actualizar(lorho): #10.49
    return np.exp(lorho-np.tile(logsumexp(lorho.T,axis=1),(K,1))).T
    
    
### Momentos de la distribucion

def Nk(R): # 10.51
    return (R.sum(axis=0))
 
def barX(R): # 10.52
    Nvec= Nk(R)
    return np.dot(matriz.T,R)/Nvec

def Sk(R): ### 10.53
    Nvec =Nk(R)
    Sksol = []
    ## Se que uso un for, pero es sobre el numero de gaussianas, y por dentro está muy optimizado
    barxk = barX(R)
    for i in range(K):
        weights = R[:,i].T
        Xk = np.subtract(matriz,barxk[:,i]) # x_n-x_k
        outer = np.einsum('ij...,i...->ij...', Xk,Xk)
        weightedouter = np.dot(outer.T,weights)
        Sksol.append(weightedouter/Nvec[i])
    return np.array(Sksol)

def Winv(Stens,Nve,bar):
    result = []
    W0inv = np.linalg.inv(W0)
    for i in range(K):   
        Wk = W0inv+Nve[i]*Stens[i]+(b0*Nve[i]/(b0+Nve[i]))*np.outer(bar[:,i]-m0,bar[:,i]-m0)
        result.append(np.linalg.inv(Wk))
    return np.array(result)

def Q_Z(Z,rho):
    return np.prod(rho**Z)
def Q_pi(pi,alpha):
    return dirichlet.pdf(pi,alpha)
def B(Wt,vt):
    return (np.linalg.det(Wt)**(-vt/3))/( (2**(vt*D/2))*(np.pi(D*(D-1)/4))*(np.prod([phi(0.5*(vt+1-(i+1))) for i in range(D)]))  )
def Wishart(lambd,Wt,vt):
    return B(wt,vt)*(np.linalg.det(lamnd)**((v-D-1)/2))*np.exp(-0.5*np.trace(np.dot(np.linalg.inv(Wt),lambd)))
```

```python
## Inferencia Variacional: Mean Fields

def VariationalInference(maxtol,ri):
    ract = ri
    for i in range(10):
        ### Calcular los momentos:   
        Ni = Nk(ract)
        barXi = barX(ract)
        Ski = Sk(ract)

        ## Actualizamos los hiperparametros
       
        alphai = a0+Ni
        betai = b0+Ni
        vi = v0+Ni

        # Distribución para los Pi
        
        Q_pi = lambda pi : dirichlet.pdf(pi,alphai)
        
        ## Distribucion para los muk y lambdak
        
        mui = (b0*np.tile(m0,(K,1)).T+Ni*barXi)/betai

        Wk = Winv(Ski,Ni,barXi)
        
        # Calculo para muk, lambda k
        
  #      for i in range(K):
  #          Q_uk_lambdak = lambda mu,lambd : multivariate_normal.pdf(mu,mean=mui[i],cov = np.linalg.inv(betai[i]*lambd)) * wishart.pdf(lambd,Wk[i],vi)
        
        ## Axtualizaciones de rho y responsabilidades
        
        logrhoi = ln_rho(alphai,betai,vi,Wk,mui)
        ract = R_actualizar(logrhoi)
    return ract
```

# Lower Bound para parar la iteración:

En el paper de Bishop, se establece que un lower bound para una KL-Divergencia para este caso particular está dado por:

$$ L = \mathbb{E}\ln (p(X|Z,\mu,\Lambda))+\mathbb{E}(\ln p(Z|\pi))+\mathbb{E}(\ln p(\pi) )+\mathbb{E}(\ln p(\mu,\Lambda) ) \\
-\mathbb{E}(\ln q(Z) )-\mathbb{E}(\ln q(\pi) )-\mathbb{E}(\ln q(\mu,\Lambda) )
$$
donde la elección de $Q$ es óptima, y cada esperanza (en el caso de $Q$) está siendo tomada con respecto al resto de las variables, como es habitual en el desarrollo de Mean-Field.

El cálculo de las esperanzas anteriores es:
1. $\mathbb{E}\ln (p(X|Z,\mu,\Lambda)) = 0.5\cdot \sum_{k=1}^KN_k [\ln |\hat{\Lambda_k}| -D\beta_k^{-1}-v_kTr(S_kW_k)-v_k(\bar{x}_k-m_k)^TW_k(\bar{x}_k-m_k)-D\ln(2\pi)]$
2. $\mathbb{E}(\ln p(Z|\pi)) = \sum_{n=1}^N \sum_{k=1}^Kr_{nk}\ln \hat{\pi} $
3. $\mathbb{E}(\ln p(\pi) ) = \ln C(\alpha_0)+(\alpha_0-1)\sum_{k=1}^K\ln\hat{\pi}$
